```python
# import the json library
import json
```

```python
# open the geojson file
with open('../data/neighborhoods-v6.geojson', 'r') as f:
    data = json.load(f)
```

```python
# create a list of objectIDs, 1 to 318
objectID_lst = list(range(1, 319))
```

```python
# add objectID to each feature's property list
index = 0
for feat in data['features']:
    feat['properties']['objectID'] = objectID_lst[index]
    print(feat['properties']['objectID'])
    index += 1
```

```python
for feat in data['features']:
    print(feat['properties'])
    print(" ")
```

```python
# write modified geojson to a new file
with open('neighborhoods-v6-mod.geojson', 'w') as f:
    json.dump(data, f)
```
